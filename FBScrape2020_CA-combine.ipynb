```python
dat = pd.read_csv(r"H:\Dropbox\FB Scrape\2020 Election Data\sample-with-content-analysis.csv", encoding = 'utf-8')
dat['Message'][82]
```

```python
print(list(dat.keys()))
```

```python
content_cats = ['Page Name', 'Trump', 'Reps', 'Biden', 'Dems', 'Public Figures', "Gov't Officials", 'Horserace', 'Campaign Event', 'Debates', 'Immigration', 'Religion', 'Biden Positive', 'Biden Negative', 'Trump Positive', 'Trump Negative', 'Threats to Democracy', 'Fact Check', 'Sex/Gender', 'Conventions', 'Terrorism', 'Military', 'Race', 'Environment', 'Crime', 'Healthcare', 'Economy', 'Covid', 'Biden Policy', 'Trump Policy', 'Justice System', 'Election Engagement', 'Emot Punctuation', 'Emot Content', 'Depict Trump', 'Depict Biden', 'American symbolism', 'Photos of a crowd', 'Campaign Materials', 'News Media Outlets', 'Uniformed Officers', 'Health Infrastructure', 'Visual Data', 'Broken Link']
```

```python
dat.groupby('Page Name').count()['pagen']
```

```python
dat[content_cats].groupby('Page Name').sum()
```

```python
for i in dat[content_cats].groupby('Page Name').sum():
    print(dat[content_cats].groupby('Page Name').sum()[i], dat[content_cats].groupby('Page Name').sum()[i] / dat.groupby('Page Name').count()['pagen'])
```

```python
category_counts = dat[content_cats].groupby('Page Name').sum()
category_counts
```

```python
category_counts.to_csv(r"H:\Dropbox\FB Scrape\2020 Election Data\Results\category_counts.csv", index = True)
```

```python
page_sums = dat.groupby('Page Name').count()['pagen']
page_sums
```

```python
category_counts['Trump'] / page_sums
```

```python
category_percents = pd.DataFrame()
for i in category_counts:
    #print(i)
    #print(category_counts[i] / page_sums)
    category_percents['{}_percent'.format(i)] = category_counts[i] / page_sums * 100
```

```python
category_percents
```

```python
category_percents.to_csv(r"H:\Dropbox\FB Scrape\2020 Election Data\Results\category_percents.csv", index = True)
```

```python
from scipy import stats as st
```

```python
for i in category_counts:
    print(st.chisquare(category_counts[i], page_sums))
```
